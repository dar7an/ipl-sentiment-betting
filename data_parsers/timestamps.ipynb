```python
import pandas as pd
import os
import transformers
import torch
import json
import pickle
```

```python
def findPath(relative_path):
    current_dir = os.getcwd()
    parent_dir = os.path.dirname(current_dir)
    target_path = os.path.join(parent_dir, relative_path)
    return target_path
```

```python
def readTimestampsOdds(filename):

    relative_path = os.path.join('the_odds_api', '2024', filename)
    with open(findPath(relative_path), 'r') as file:
        data = json.load(file)


    last_updates = []
    for log in data:
        # Extracting the "last_update" information for odds
        last_update = log['odds']['data']['bookmakers'][0]['markets'][0]['last_update']
        last_updates.append(last_update)

    return last_updates

print(readTimestampsOdds('1.json'))
```

```python
relative_path = os.path.join('the_odds_api', '2024')
odds_files = os.listdir(findPath(relative_path))


timestamps = {}
for filename in odds_files:
    try:
        timestampsOdds = readTimestampsOdds(filename)
        timestamps[filename.replace(".json", "")] = timestampsOdds
    except Exception:
        print(f"Exception: Error in processing {filename}")
    

print(timestamps)
```

```python
# Pickle the dictionary
with open('timestamps.pkl', 'wb') as file:
    pickle.dump(timestamps, file)
```
